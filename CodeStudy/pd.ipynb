```python
left = pd.DataFrame(
    {"key1": ["foo", "foo", "bar"], "key2": ["one", "two", "one"], "lval": [1, 2, 3]}
)
right = pd.DataFrame(
    {
        "key1": ["foo", "foo", "bar", "bar"],
        "key2": ["one", "one", "one", "two"],
        "rval": [4, 5, 6, 7],
    }
)
df1 = pd.merge(
    left, right, on="key1", how="outer", suffixes=["_left", "_right"]
).fillna(method="ffill")
df2 = pd.merge(left, right, left_index=True, right_index=True, how="inner")
```
