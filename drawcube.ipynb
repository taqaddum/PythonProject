```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib widget
```

```python
x,y,z = np.indices((3, 2, 3))
ax = plt.subplot(projection='3d')
filled = np.ones((2, 1, 2), dtype=bool)
ax.voxels(x, y, z, filled=filled, edgecolors='black', lw=0.5, facecolor=[1, 1, 1, 0.1])
ax.plot3D([0.5, 1.5, 1.5, 0.5, 0.5], [0.5, 0.5, 0.5, 0.5, 0.5], [1.5, 1.5, 0.5, 0.5, 1.5], lw=3)
ax.set_axis_off()
plt.show()
```
